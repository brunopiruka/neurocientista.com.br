#Panorama da **COVID-19**

Apesar de toda a "fama" recente, o coronavírus, ou melhor, os coronavírus contituem uma classe de vírus já conhecida há muito tempo. Com toda certeza, em algum momento de sua vida, você já foi "vítima" de algum tipo de coronavírus. Eles são responsáveis por causarem os chamados resfriados comuns.

Então o que a **COVID-19** tem de diferente?

Não à toa, vem sendo chamado de **novo coronavírus**, justamente por apresentar uma diferença marcante. Seu grande poder de propagação. Dessa forma, apesar da baixa letalidade, é uma enorme ameaça pela rapidez e facilidade com a qual se prolifera.

##Compreendendo as Variáveis*

Antes de iniciar a análise, cabe elucidar o significado das variáveis. Dessa forma, segue um breve dicionário que ajudará a guiar os próximos passos:

* `iso_code` - Identificação do País segundo padrão internacional de 3 Letras
* `continent` - Continente o qual perntence o País
* `location` - Localização Geográfica
* `date`- Data da Observação
* `total_cases` - Número total de casos confirmados de **COVID-19**
* `new_cases`- Número de novos casos de **COVID-19**
* `total_deaths`- Número total de mortes por **COVID-19**
* `new_deaths` - Novas mortes por **COVID-19**
* `total_cases_per_million` - Total de casos confirmados por 1 Milhão de pessoas
* `new_cases_per_million` - Novo casos confirmados por 1 Milhão de pessoas
* `total_deaths_per_million` - Total de mortes por 1 Milhão de pessoas
* `new_deaths_per_million` - Novas mortes por 1 Milhão de pessoas
* `total_tests` - Total de testes realizados
* `new_tests` - Novos testes realizados
* `new_tests_smoothed` - Novos testes realizados, ajustados para 7 dias, para países que não reportam o número diário de novos testes
* `total_tests_per_thousand` - Total de testes a cada 1.000 pessoas
* `new_tests_per_thousand` - Novos testes a cada 1.000 pessoas
* `new_tests_smoothed_per_thousand` - Novos testes, ajustados para 7 dias, a cada 1.000 pessoas
* `tests_units` - Unidades utilizadas pelo local para reportar seus dados de testagem
* `stringency_index` - Índice de resposta governamental, em uma escala de 0 a 100, baseada em 9 indicadores, que envolvem medidas restritivas como fechamento de escolas, comércio, proibição de viagens, etc.
* `population` - População em 2020
* `population_density` - Densidade populacional, medida em quilômetros quadrados, para o ano mais recente disponível
* `median_age`- Idade média da população, projetada para 2020
* `aged_65_older` - Parcela da população com 65 anos ou mais
* `aged_70_older` - Parcela da população com 70 anos ou mais
* `gdp_per_capita` - Produto Interno Bruto, com paridade de poder de compra (Dólares internacionais de 2011), para o ano mais recente disponível
* `extreme_poverty` - Parcela da população que vive em situação de extrema pobreza, para o ano mais recente disponível
* `cvd_death_rate` - Taxa de mortalidade por doenças cardiovasculares em 2017
* `diabetes_prevalence` - Prevalência de diabetes (% da população com idade entre 29 e 70 anos) em 2017
* `female_smokers` - Parcela de mulheres fumantes, para o ano mais recente disponível
* `male_smokers` - Parcela de homens fumantes, para o ano mais recente disponível
* `handwashing_facilities` - Parcela da população com acesso a formas de higienização das mãos, para o ano mais recente disponível
* `hospital_beds_per_thousand` - Leitos de hospitais para cada 1.000 pessoas, para o ano mais recente disponível desde 2010
* `life_expectancy` - Epectativa de vida para nascidos em 2019

##Importando os Dados

Após a breve introdução e explicação, vamos importar os dados e iniciar nossas análises

```
# Importando as bilbiotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Impotando o dataset e criando nosso dataframe

dataset = ('https://covid.ourworldindata.org/data/owid-covid-data.csv')
df = pd.read_csv(dataset)
```

##Análise Exploratória dos Dados

Como é possível perceber, o *dataset* nos apresenta inúmeras informações. Porém, para nossa análise, nos ateremos principalmente ao número de casos e número de mortes.

###Análise Prévia do *Dataset*

* Sempre é recomendável análisar as primeiras entradas do *dataset* no intuito de ter uma primeira impressão das características dos dados que iremos trabalhar

```
# Visualizando as 5 primeiras entradas

df.head()
```

* O segundo passo é ter a noção do tamanho do *dataset* com o qual estamos trabalhando

```
# Descobrindo o tamanho do dataset

print('Número de linhas: ', df.shape[0])
print('Número de colunas: ', df.shape[1])
```

* Vamos agora identificar o nome das colunas e os tipos de variáveis de cada uma delas


```
# Identificando o nome das colunas

df.dtypes
```

* Já é possível perceber que a columa `date` está em formato `object`. Portanto, iremos transformá-la em tipo `data`

```
df.date = pd.to_datetime(df.date)
df.dtypes
```

* Uma vez que já conhecemos melhor nosso *dataset* de trabalho é prudente avaliar possíveis valores ausentes e a implicância para a análise

```
# Verificando o percentual de valores ausentes

(df.isnull().sum().sort_values(ascending=False) / df.shape[0]).round(2)
```

* Agora que já transformamos a columa `date` para o tipo `data` podemos realizar operações, como por exemplo localizar os países com o maior número de casos e mortes para a data atual

```
# Localizando os países com maiores números de casos hoje
from datetime import date
hoje = date.today().strftime('%Y-%m-%d')
df.loc[df.date == hoje ].sort_values(by='total_cases', ascending=False)
```

```
# Localizando os países com maiores números de mortes hoje
from datetime import date
hoje = date.today().strftime('%Y-%m-%d')
df.loc[df.date == hoje ].sort_values(by='total_deaths', ascending=False)
```

* Como forma de visualizar melhor os dados, podemos plotar um gráfico para os 5 países com maior número de mortes


```
# Plontando um gráfico dos 5 países com maior número de mortes

maiores_mortes = df.loc[df.date == hoje, ['location','total_deaths']].sort_values(by='total_deaths', ascending=False)[1:6]

sns.barplot(x='location', y='total_deaths', data=maiores_mortes)

plt.tight_layout()
```

* Podemos também avaliar a evolução mundial da **COVID-19** no mundo, através de um gráfico de linha

```
# Plotando um gráfico de linha para a evolução no mundo

mundo = df.loc[df.location == 'World', 'total_cases'].reset_index(drop=True)

fig, ax = plt.subplots()
ax.set_title('Evolução da COVID-19 no mundo')
ax.set_ylabel('Número de Casos')
ax.set_xlabel('Evolução em Dias')
ax.plot(mundo)

plt.tight_layout()

```

* Podemos, por exemplo, tentar analisar alguns outros fatores, tais como PIB per capita e número de mortes

```
# Plotando 
gdp = df.loc[df.date == hoje, ['total_deaths','gdp_per_capita']]
plt.scatter(x='gdp_per_capita', y='total_deaths', data=gdp)
plt.xlabel('PIB per Capita')
plt.ylabel('Número de Mortes')
plt.legend()

plt.tight_layout()

```

#**COVID-19** no Brasil

Agora podemos avaliar especificamente a evolução da pandemia apenas no Brasil. Para a análise que segue, criaremos uma cópia do dataframe, porém, apenas com os dados do Brasil

```
# Criando uma cópia do dataframe apenas para o Brasil

df_brasil = df.loc[df.location == 'Brazil'].copy()

# Visualizando as primieiras entradas
df_brasil.head()
```

* Com base nesses dados iremos encontrar a data do primeiro caso de **COVID-19** no Brasil

```
# Encontrando o registro do primeiro caso de COVID-19 no Brasil

primeiro_caso = df_brasil.loc[df_brasil.total_cases == 1]
print('O Primeiro Caso no Brasil foi em {}'.format(primeiro_caso.date))
```

```
data_primeiro_caso = primeiro_caso.date.loc[3694]
```

* Com base nesses dados iremos encontrar a data da primeira morte por **COVID-19** no Brasil

```
# Encontrando o registro da primeira morte por COVID-19 no Brasil

primeira_morte = df_brasil.loc[df_brasil.total_deaths == 1]
print('A Primeira Morte no Brasil foi em {}'.format(primeira_morte.date))
```

```
data_primeira_morte = primeira_morte.date.loc[3715]
```

* Dessa forma é possível calcular o tempo decorrido entre o registro do primeiro caso e da primeira morte por **COVID-19** no Brasil

```
# Calculando o tempo decorrido entre o primeiro caso e a primeira morte

tempo = data_primeira_morte - data_primeiro_caso
print('O tempo entre o primeiro caso e a primeira morte foi de {} dias'.format(tempo.days))
```

* Para finalizar podemos plotar os gráficos e avaliar a evolução da doença - casos totais e mortes totais - de forma linear e logarítmica

```
#Plotando os gráficos lineares e logarítmicos

# Limpando o dataset para iniciar a análise partindo do primeiro caso confirmado
index_casos = df_brasil[df_brasil.total_cases == 0.0 ].index
df_brasil.drop(index_casos, inplace=True)

# Reiniciando o índice para contar partindo do primeiro caso confirmado
casos_brasil = df_brasil.total_cases.reset_index(drop=True)
mortes_brasil = df_brasil.total_deaths.reset_index(drop=True)

#Estruturando os gráficos
fig, ((axc,axcl),(axm,axml)) = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
axc.set_title('Evolução de Casos no Brasil - Linear')
axc.set_ylabel('Número de Casos')
axc.set_xlabel('Evolução em Dias')
axc.plot(casos_brasil)

axcl.set_title('Evolução de Casos no Brasil - Logarítmica')
axcl.set_ylabel('Número de Casos')
axcl.set_yscale('log')
axcl.set_xlabel('Evolução em Dias')
axcl.plot(casos_brasil)

axm.set_title('Evolução de Mortes no Brasil - Linear')
axm.set_ylabel('Número de Casos')
axm.set_xlabel('Evolução em Dias')
axm.plot(mortes_brasil)

axml.set_title('Evolução de Mortes no Brasil - Logarítmica')
axml.set_ylabel('Número de Casos')
axml.set_yscale('log')
axml.set_xlabel('Evolução em Dias')
axml.plot(mortes_brasil)

plt.tight_layout()
```
